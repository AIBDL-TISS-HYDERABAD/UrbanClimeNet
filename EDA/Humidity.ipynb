```python
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
```

```python
df = pd.read_csv('C:/Users/TripleA/Desktop/coding related/python/TISS/1_MONTH.csv', encoding='unicode_escape')
```

```python
df.dropna(inplace=True)
df.dtypes
```

```python
df.shape
```

```python
df['Date & Time'] = pd.to_datetime(df['Date & Time'], format ='mixed' )
```

```python
df.dtypes
```

```python
df.set_index('Date & Time' , inplace=True)
```

```python
df.head()
```

HUMIDITY

    WEEK

```python
per_week = df[['Hum - %', 'High Hum - %','Low Hum - %']].resample('W')
```

```python
plt.plot(per_week['High Hum - %'].max(),'.-', label = 'MAX')
plt.plot(per_week['Hum - %'].mean(),'.-', label = 'AVG')
plt.plot(per_week['Low Hum - %'].min(),'.-', label = 'MIN')
plt.xticks(rotation=45)
plt.ylabel('Hum - %')
plt.title('Humidity in % per week')
plt.grid(True)
plt.legend()
plt.gcf().autofmt_xdate()
```

PER DAY

```python
per_day = df[['Hum - %', 'High Hum - %','Low Hum - %']].resample('D')
```

```python
plt.plot(per_day['High Hum - %'].max(),'.-', label = 'MAX')
plt.plot(per_day['Hum - %'].mean(),'.-', label = 'AVG')
plt.plot(per_day['Low Hum - %'].min(),'.-', label = 'MIN')
plt.xticks(rotation=45)
plt.ylabel('Hum in %')
plt.title('Hum in % per day')
plt.grid(True)
plt.legend()
```
