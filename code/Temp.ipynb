```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/TripleA/Desktop/coding related/python/TISS/1_MONTH.csv' , encoding='unicode_escape')
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df['Date & Time'] = pd.to_datetime(df['Date & Time'] , format= 'mixed')
```

```python
df.dtypes
```

```python
df.set_index('Date & Time',inplace=True)
```

TEMPERATURE

         PER WEEK

```python
per_week = df[['Temp - °C','High Temp - °C', 'Low Temp - °C']].resample('W')
```

```python
plt.plot(per_week['Temp - °C'].mean(),'.-', label = 'AVG')
plt.plot(per_week['High Temp - °C'].max(),'.-', label = 'MAX')
plt.plot(per_week['Low Temp - °C'].min(),'.-', label = 'MIN')
plt.xticks(rotation=45)
plt.ylabel('Temp in °C ')
plt.title('Temperature per week')
plt.grid(True)
plt.legend()
plt.gcf().autofmt_xdate()
```

          PER DAY

```python
per_day = df[['Temp - °C','High Temp - °C', 'Low Temp - °C']].resample('D')
```

```python
plt.plot(per_day['High Temp - °C'].max(),'.-', label = 'MAX')
plt.plot(per_day['Temp - °C'].mean(),'.-', label = 'AVG')
plt.plot(per_day['Low Temp - °C'].min(),'.-', label = 'MIN')
plt.xticks(rotation=45)
plt.ylabel('Temp in °C ')
plt.title('Temperature per day')
plt.grid(True)
plt.legend()
```
